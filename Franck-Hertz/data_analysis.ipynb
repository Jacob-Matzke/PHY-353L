```python
import glob
import os
import pandas as pd

def find_all_directories():
    dirs = [p for p in glob.glob("Mercury Data/*") if os.path.isdir(p)]
    return dirs

def load_csv_to_df(filename):
    return pd.read_csv(filename, skiprows=11)

def combine_channels(x_df, y_df):
    return pd.DataFrame({
        "Second": x_df["Second"],
        "V": x_df["Value"],
        "I": y_df["Value"]
    })
    
import matplotlib.pyplot as plt

def graph_value1_against_value2(df):
    plt.figure()
    plt.scatter(df["V"], df["I"], s=0.5)
    plt.xlabel("U_2/V")
    plt.ylabel("I/nA")
    plt.title("V vs I")
    plt.grid(True)
    plt.show()
```

```python
path = "Mercury Data\\Run 1"
channel_1_fn = "SDS824X_HD_CSV_C1_1.csv"
channel_2_fn = "SDS824X_HD_CSV_C2_1.csv"

path_1 = os.path.join(path, channel_1_fn)
path_2 = os.path.join(path, channel_2_fn)

df_1 = load_csv_to_df(path_1)
df_2 = load_csv_to_df(path_2)

combined = combine_channels(df_1, df_2)

graph_value1_against_value2(combined)
```
